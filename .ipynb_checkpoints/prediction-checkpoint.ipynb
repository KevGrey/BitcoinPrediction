```python
import yfinance as yf #连接雅虎经济接口
import os
import pandas as pd
```

```python
btc_ticker = yf.Ticker("BTC-USD") #下载比特币对应美元历史价格
```

```python
if os.path.exists("btc.csv"):
    btc = pd.read_csv("btc.csv", index_col=0)
else:
    btc = btc_ticker.history(period="max")
    btc.to_csv("btc.csv")
```

```python
btc
```

```python
btc.index = pd.to_datetime(btc.index)
```

```python
del btc["Dividends"]
del btc["Stock Splits"]
```

```python
btc.columns = [c.lower() for c in btc.columns]
#转换列名为小写
```

```python
btc.plot.line(y="close", use_index=True)
```

```python
wiki = pd.read_csv("wikipedia_edits.csv", index_col=0, parse_dates=True)
```

```python
wiki
```

```python
btc = btc.merge(wiki, left_index=True, right_index=True)
#合并数据
```

```python
btc["tomorrow"] = btc["close"].shift(-1)
#明日价格设为下一天实际价格
```

```python
btc["target"] = (btc["tomorrow"] > btc["close"]).astype(int)
#1代表涨，0代表跌
```

```python
btc["target"].value_counts()
#做回归前检查训练集
```

```python
btc
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=50, random_state=1)

#RandomForest作基本模型，因为不会过拟合,estimators决策树数量，
#min samoles split每棵树最少50个开始分枝（数量过大会使训练集效果下降，过小会过拟合），
#random state=1使得不改变数据时每次结果相同
train = btc.iloc[:-365]
#去掉最后200天的数据作为训练集
test = btc.iloc[-365:]
#最后200天数据作为测试集

predictors = ["close", "volume", "open", "high", "low", "edit_count", "sentiment", "neg_sentiment"]


#设置预测因子
model.fit(train[predictors], train["target"])
```

```python
from sklearn.metrics import precision_score
#检测初次精确率
preds = model.predict(test[predictors])
preds = pd.Series(preds, index=test.index)
precision_score(test["target"], preds)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="predictions")
    combined = pd.concat([test["target"], preds], axis=1)#将预测的值和真实值合并在一个表格内
    return combined
```

```python
#backtest可以用时期相隔大的不同时间段评估预测结果
def backtest(data, model, predictors, start=1095, step=150):#start是预测日期前的天数，设置为三年，step指每150天（半年）预测一次
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions)
```

```python
#xgboost分别训练不同独立模型的决策树,将一个结果和另一个结果联系
from xgboost import XGBClassifier

model = XGBClassifier(random_state=1, learning_rate=.1, n_estimators=200)
predictions = backtest(btc, model, predictors)
```

```python
predictions["predictions"].value_counts()
```

```python
precision_score(predictions["target"], predictions["predictions"])
```

```python
def compute_rolling(btc):
    horizons = [2,7,60,365]#分别通过计算2天，7天，60天，365天的趋势，并作为新因素加入
    new_predictors = ["close", "sentiment", "neg_sentiment"]

    for horizon in horizons:
        rolling_averages = btc.rolling(horizon, min_periods=1).mean()

        ratio_column = f"close_ratio_{horizon}"
        btc[ratio_column] = btc["close"] / rolling_averages["close"]
        
        edit_column = f"edit_{horizon}"
        btc[edit_column] = rolling_averages["edit_count"]

        rolling = btc.rolling(horizon, closed='left', min_periods=1).mean()
        trend_column = f"trend_{horizon}"
        btc[trend_column] = rolling["target"]

        new_predictors+= [ratio_column, trend_column, edit_column]
    return btc, new_predictors
```

```python
btc
```

```python
btc, new_predictors = compute_rolling(btc.copy())
```

```python
predictions = backtest(btc, model, new_predictors)
```

```python
precision_score(predictions["target"], predictions["predictions"])
```
